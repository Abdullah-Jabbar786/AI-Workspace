```python
my_dict = {
    "name" : "Abdullah",
    "marks" : 90,
    "grade" : ""}

print(my_dict)

if(my_dict["marks"] >= 90):
    my_dict["grade"] = "A+"
elif(my_dict["marks"] >= 80):
    my_dict["grade"] = "A"
elif(my_dict["marks"] >= 70):
    my_dict["grade"] = "B"
elif(my_dict["marks"] >= 60):
    my_dict["grade"] = "C"
elif(my_dict["marks"] >= 50):
    my_dict["grade"] = "D"
else: my_dict["grade"] = "E"

print(my_dict)
```

```python
import math

num = int(input("Enter a number: "))
isPrime = True

if num<=1:
    isPrime = False
else:
    for i in range(2, int(math.sqrt(num)) + 1):
        if num%i == 0:
            isPrime = False
            break

print(isPrime)
```

```python
def find_max(*args):
    return max(args)

def print_key_values(**kwargs):
    for key, value in kwargs.items():
        print(f"{key} : {value}")

print(find_max(10, 25, 5, 40, 15))

print_key_values(name="Abdullah", age=20, city="New York")
```
