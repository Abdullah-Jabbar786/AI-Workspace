```python
class BankAccount:
    bank_name = "AHM Bank" 
    def __init__(self, holder_name, account_no, balance):
        self.holder_name = holder_name 
        self.account_no = account_no 
        self.balance = balance 
    def display_account(self): # info
        print("Bank:", BankAccount.bank_name)
        print("Account Holder:", self.holder_name)
        print("Account Number:", self.account_no)
        print("Balance:", self.balance)

    def deposit(self, amount): # method to deposit money
        self.balance += amount
        print(f"Deposited {amount}. New Balance: {self.balance}")

account1 = BankAccount("Ali", 1001, 5000)
account2 = BankAccount("Sara", 1002, 7000)

account1.display_account()
print("----------------")
account2.display_account()
print("----------------")
account1.deposit(2000)
```

```python
class SmartFan:
    def __init__(self, room_name, speed):
        self.room_name = room_name
        self.speed = 0

    def increase_speed(self, x):
        print(f"Increasing speed of Fan by {x}")
        self.speed += x
        
    def decrease_speed(self, x):
        print(f"Decreasing speed of Fan by {x}")
        self.speed -= x

    def display_status(self): # info
        print("Room Name: ", self.room_name)
        print("Fan Speed: ", self.speed)

sf1 = SmartFan("Room R1", 85)
sf2 = SmartFan("Room R2", 75)


sf1.display_status()
sf1.increase_speed(15)
sf1.decrease_speed(5)
sf1.display_status()
sf2.display_status()
sf2.increase_speed(25)
sf2.decrease_speed(5)
sf2.display_status()
```

```python
class Book:
    library_name = "Fast Library"

    def __init__(self, title, author):
        self.title = title
        self.author = author

    def display_book(self):
        print(f"Library Name: {self.library_name}")
        print(f"Title: {self.title}")
        print(f"Title: {self.author}")
        print("-"*20)

B1 = Book("The Republic","Plato")
B2 = Book("Meditations", "Marcus Aurelius")

print("Before Change: ")
B1.display_book()
B2.display_book()

# Changing the Class Attribute
Book.library_name = "Digital AI Library"

print("After Change: ")
B1.display_book()
B2.display_book()
```

```python
class Bank:
    def __init__(self, account_holder, account_no, balance):
        self.account_holder = account_holder
        self.account_no = account_no
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount
        print(f"Deposited {amount}. New Balance: {self.balance}")
        
    def withdraw(self, amount):
        if amount<=self.balance:
            self.balance -= amount
            print(f"Withdrew {amount}. New Balance: {self.balance}")
        else:
            print("Insufficient funds! \nNO Transaction Occured\n")
            
    def display_account(self):
        print(f"Account Holder Name: {self.account_holder}")
        print(f"Account Number: {self.account_no}")
        print(f"Account Balance: {self.balance}")
        print("-"*20)

Acc1 = Bank("Abdullah", 859, 1000000)
Acc2 = Bank("Hassaan", 1031, 800000)

Acc1.deposit(50000)
Acc1.withdraw(20000)
Acc1.display_account()

Acc2.withdraw(900000) 
Acc2.display_account()
```

```python
class SmartLight:
    def __init__(self, room, status):
        self.room = room
        self.status = status

    def turn_on(self):
        self.status = True

    def turn_off(self):
        self.status = False

    def show_status(self):
        state_text = "ON" if self.status else "OFF"
        print(f"Room: {self.room}")
        print(f"Light Status: {state_text}")
        print("-" * 20)
        
sl1 = SmartLight("R1" , True)
sl1.turn_off()
sl1.show_status()

sl2 = SmartLight("R2" , False)
sl2.turn_on()
sl2.show_status()      
```

```python
class Staff:
    def __init__(self, name, staff_id, department):
        self.name = name
        self.staff_id = staff_id
        self.department = department

    def display_info(self):
        print(f"Staff Name: {self.name}")
        print(f"ID: {self.staff_id}")
        print(f"Department: {self.department}")

class Teacher(Staff):
    def __init__(self,name, staff_id, department, courses, salary):
        super().__init__(name, staff_id, department)
        self.courses = courses
        self.salary = salary

    def display_info(self):
        super().display_info()
        print(f"Courses: {self.courses}")
        print(f"Salary: {self.salary}")    

class AdminStaff(Staff):
    def __init__(self,name, staff_id, department, role, working_hours):
        super().__init__(name, staff_id, department)
        self.role = role
        self.working_hours = working_hours

    def display_info(self):
        super().display_info()
        print(f"Role: {self.role}")
        print(f"Working Hours: {self.working_hours}")    

class ResearchAssistant(Staff):
    def __init__(self,name, staff_id, department, research_topic, stipend):
        super().__init__(name, staff_id, department)
        self.research_topic = research_topic
        self.stipend = stipend

    def display_info(self):
        super().display_info()
        print(f"Research Topic: {self.research_topic}")
        print(f"Stipend: {self.stipend}")    

t1 = Teacher("Abdullah", "A859", "AI Dept", "Machine Learning", 150000)
a1 = AdminStaff("Hassaan", "H-1031", "Registrar", "Coordinator", 40)
ra1 = ResearchAssistant("Majid", "M-895", "AI Researcher", "Deep Learning", 100000)
t1.display_info()
print("-"*20)
a1.display_info()
print("-"*20)
ra1.display_info()
```

```python
class Vehicle:
    def start(self):
        print("Vehicle is starting...")

class Car(Vehicle):
    def start(self):
        print("Car: Engine revving, dashboard lights on.")

class Bike(Vehicle):
    def start(self):
        print("Bike: Kickstart engaged.")

class Bus(Vehicle):
    def start(self):
        print("Bus: Air compressor active, diesel engine rattling")

vehicles = [Car(), Bike(), Bus()] # list of objects

for v in vehicles: # loop for same function calling of all objects
    v.start()
    print("-"*20)
```

```python
class Calculator:
    def multiply(self, a, b=1):
        print("Product: ", a*b)
    def multiply(self, *args):
        result = 1
        for num in args:
            result = result * num
        return result

calc = Calculator()
print("Method Overloading default arg Examples: ")
print(calc.multiply(5, 10))
print(calc.multiply(7))
print("Method Overloading *args Examples: ")
print(calc.multiply(5, 10)) 
print(calc.multiply(5, 10, 15)) 
print(calc.multiply(1, 2, 3, 4)) 
```

```python
class BankAccount:
    def __init__(self, balance):
        self.__balance = balance 

    def deposit(self, amount):
        self.__balance += amount

    def withdraw(self, amount):
        if amount<=self.balance:
            self.__balance -= amount
            print(f"Withdrew {amount}. New Balance: {self.__balance}")
        else:
            print("Insufficient funds! \nNO Transaction Occured\n")
            
    def get_balance(self):
        print(f"Account Balance: {self.__balance}")

acc = BankAccount(10000)
acc.get_balance() # works because method is public

try:
    print(acc.__balance) 
except AttributeError:
    print("Success: Cannot access __balance directly from outside!")

acc.__balance = 50000
print("Tried to change balance...")
acc.get_balance()
```
